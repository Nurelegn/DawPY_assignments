## Functions  to fill the missing values:

### 1) first_looking(col)

```python
def first_looking(*cols):
    for col in cols:
        print("COLUMN NAME    : ", col)
        print("--------------------------------")
        print("per_of_nulls   : ", "%", round(df4[col].isnull().sum()*100 / df4.shape[0], 2))
        print("num_of_nulls   : ", df4[col].isnull().sum())
        print("num_of_uniques : ", df4[col].nunique())
        print("--------------------------------")
        print(df4[col].value_counts(dropna=False))
        print("\n")
```

### 2) fill_most_freq(df, group_col, col_name)

```python
def fill_most_freq(df4, group_col, col_name):
    for group in list(df4[group_col].unique()):
        cond = df4[group_col]==group
        mode = list(df4[cond][col_name].mode())
        if mode != []:
            df4.loc[cond, col_name] = df4.loc[cond, col_name].fillna(df[cond][col_name].mode()[0])
        else:
            df4.loc[cond, col_name] = df4.loc[cond, col_name].fillna(df4[col_name].mode()[0])
    
    print("COLUMN NAME    : ", col_name)
    print("--------------------------------")
    print("per_of_nulls   : ", "%", round(df[col_name].isnull().sum()*100 / df.shape[0], 2))
    print("num_of_nulls   : ", df4[col_name].isnull().sum())
    print("num_of_uniques : ", df4[col_name].nunique())
    print("--------------------------------")
    print(df4[col_name].value_counts(dropna = False).sort_index())
```

### 3) fill_mode(df, group_col1, group_col2, col_name)

```python
def fill_mode(df4, group_col1, group_col2, col_name):
    for group1 in list(df[group_col1].unique()):
        for group2 in list(df[group_col2].unique()):
            cond1 = df[group_col1]==group1
            cond2 = (df[group_col1]==group1) & (df[group_col2]==group2)
            mode1 = list(df[cond1][col_name].mode())
            mode2 = list(df[cond2][col_name].mode())
            if mode2 != []:
                df4.loc[cond2, col_name] = df4.loc[cond2, col_name].fillna(df4[cond2][col_name].mode()[0])
            elif mode1 != []:
                df4.loc[cond2, col_name] = df4.loc[cond2, col_name].fillna(df4[cond1][col_name].mode()[0])
            else:
                df4.loc[cond2, col_name] = df4.loc[cond2, col_name].fillna(df4[col_name].mode()[0])
    
    print("column name    : ", col_name)
    print("--------------------------------")
    print("per_of_nulls   : ", "%", round(df4[col_name].isnull().sum()*100 / df4.shape[0], 2))
    print("num_of_nulls   : ", df4[col_name].isnull().sum())
    print("num_of_uniques : ", df4[col_name].nunique())
    print("--------------------------------")
    print(df4[col_name].value_counts(dropna = False))
```

### 4) fill_prop(df, group_col, col_name)

```python
def fill_prop(df4, group_col, col_name):
    for group in list(df4[group_col].unique()):
        cond = df4[group_col]==group
        df4.loc[cond, col_name] = df4.loc[cond, col_name].fillna(method="ffill").fillna(method="bfill")
    df4[col_name] = df4[col_name].fillna(method="ffill").fillna(method="bfill")
    
    print("COLUMN NAME    : ", col_name)
    print("--------------------------------")
    print("per_of_nulls   : ", "%", round(df[col_name].isnull().sum()*100 / df4.shape[0], 2))
    print("num_of_nulls   : ", df4[col_name].isnull().sum())
    print("num_of_uniques : ", df4[col_name].nunique())
    print("--------------------------------")
    print(df4[col_name].value_counts(dropna = False).sort_index())
```

### 5) fill(df, group_col1, group_col2, col_name, method)

```python
def fill(df4, group_col1, group_col2, col_name, method):
    if method == "mode":
        for group1 in list(df4[group_col1].unique()):
            for group2 in list(df4[group_col2].unique()):
                cond1 = df4[group_col1]==group1
                cond2 = (df4[group_col1]==group1) & (df4[group_col2]==group2)
                mode1 = list(df4[cond1][col_name].mode())
                mode2 = list(df4[cond2][col_name].mode())
                if mode2 != []:
                    df4.loc[cond2, col_name] = df4.loc[cond2, col_name].fillna(df4[cond2][col_name].mode()[0])
                elif mode1 != []:
                    df4.loc[cond2, col_name] = df4.loc[cond2, col_name].fillna(df4[cond1][col_name].mode()[0])
                else:
                    df4.loc[cond2, col_name] = df4.loc[cond2, col_name].fillna(df4[col_name].mode()[0])

    elif method == "mean":
        df4[col_name].fillna(df4.groupby([group_col1, group_col2])[col_name].transform("mean"), inplace = True)
        df4[col_name].fillna(df4.groupby(group_col1)[col_name].transform("mean"), inplace = True)
        df4[col_name].fillna(df4[col_name].mean(), inplace = True)
        
    elif method == "median":
        df4[col_name].fillna(df4.groupby([group_col1, group_col2])[col_name].transform("median"), inplace = True)
        df4[col_name].fillna(df4.groupby(group_col1)[col_name].transform("median"), inplace = True)
        df4[col_name].fillna(df4[col_name].median(), inplace = True)
        
    elif method == "ffill":           
        for group1 in list(df4[group_col1].unique()):
            for group2 in list(df4[group_col2].unique()):
                cond2 = (df4[group_col1]==group1) & (df4[group_col2]==group2)
                df4.loc[cond2, col_name] = df4.loc[cond2, col_name].fillna(method="ffill").fillna(method="bfill")
                
        for group1 in list(df4[group_col1].unique()):
            cond1 = df4[group_col1]==group1
            df4.loc[cond1, col_name] = df4.loc[cond1, col_name].fillna(method="ffill").fillna(method="bfill")            
           
        df4[col_name] = df4[col_name].fillna(method="ffill").fillna(method="bfill")
    
    print("COLUMN NAME    : ", col_name)
    print("--------------------------------")
    print("per_of_nulls   : ", "%", round(df4[col_name].isnull().sum()*100 / df4.shape[0], 2))
    print("num_of_nulls   : ", df4[col_name].isnull().sum())
    print("num_of_uniques : ", df4[col_name].nunique())
    print("--------------------------------")
    print(df4[col_name].value_counts(dropna = False).sort_index())
```

```python
#  applying user defined function for imputing missing values in "co_emissions"

def fill_median(df4, group_col1, group_col2, group_col3, col_name):
    
    df4[col_name].fillna(df4.groupby([group_col1, group_col2, group_col3])[col_name].transform("median"), inplace = True)
    df4[col_name].fillna(df4.groupby([group_col1, group_col2])[col_name].transform("median"), inplace = True)
    df4[col_name].fillna(df4.groupby(group_col1)[col_name].transform("median"), inplace = True)
    df4[col_name].fillna(df4[col_name].median(), inplace = True)

    print("column name    : ", col_name)
    print("--------------------------------")
    print("per_of_nulls   : ", "%", round(df4[col_name].isnull().sum()*100 / df4.shape[0], 2))
    print("num_of_nulls   : ", df4[col_name].isnull().sum())
    print("num_of_uniques : ", df4[col_name].nunique())
    print("--------------------------------")
    print(df4[col_name].value_counts(dropna = False))
```

```python
# This DEF will fill the missing values in the "co_emissions" with the median value based on "make_model", "body_type", and "fuel_type"

fill_median(df4, "make_model", "body_type", "fuel_type", "co_emissions")
```
